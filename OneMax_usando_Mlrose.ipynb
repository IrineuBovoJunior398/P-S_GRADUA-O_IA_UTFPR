```python
# Configurações para execução do Framework
import six
import sys
sys.modules['sklearn.externals.six'] = six
import mlrose
```

```python
# Construção da função fitness
# objective function
def fitness_function(individual):
  # The sum of all ones in the individual (list)
  count = 0
  for bit in individual:
    count += bit
  return count
```

```python
fitness = mlrose.CustomFitness(fitness_function)
```

```python
problema = mlrose.DiscreteOpt(length = 5, fitness_fn = fitness, maximize = True, max_val = 2)
```

```python
# ALGORITMO GENÉTICO
solucao_GA, melhor_custo = mlrose.genetic_alg(problema, pop_size=8, mutation_prob=0.2, max_iters=4)
solucao_GA, melhor_custo
```
