```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Preços esperados
actual_prices = np.array([270000, 360000, 410000, 500000, 550000])

# Predições do Cenário 1
predicted_prices_scenario1 = np.array([390000, 430000, 635000, 495000, 590000])

# Predições do Cenário 2
predicted_prices_scenario2 = np.array([310000, 390000, 480000, 570000, 660000])

# Calculando as medidas do Cenário 1
mae_scenario1 = mean_absolute_error(actual_prices, predicted_prices_scenario1)
mse_scenario1 = mean_squared_error(actual_prices, predicted_prices_scenario1)
rmse_scenario1 = np.sqrt(mse_scenario1)
r2_scenario1 = r2_score(actual_prices, predicted_prices_scenario1)

# Calculando as medidas do Cenário 2
mae_scenario2 = mean_absolute_error(actual_prices, predicted_prices_scenario2)
mse_scenario2 = mean_squared_error(actual_prices, predicted_prices_scenario2)
rmse_scenario2 = np.sqrt(mse_scenario2)
r2_scenario2 = r2_score(actual_prices, predicted_prices_scenario2)

# Mostrando os resultados
print("Medidas para Cenário 1:")
print("MAE:", mae_scenario1)
print("MSE:", mse_scenario1)
print("RMSE:", rmse_scenario1)
print("R²:", r2_scenario1)
print("Medidas para Cenário 2:")
print("MAE:", mae_scenario2)
print("MSE:", mse_scenario2)
print("RMSE:", rmse_scenario2)
print("R²:", r2_scenario2)
```

```python
import matplotlib.pyplot as plt

# Plotando
plt.figure(figsize=(10, 6))

# Scatter para o Cenário 1
plt.scatter(actual_prices, predicted_prices_scenario1, color='blue', label='Scenário 1')

# Scatter para o Cenário 2
plt.scatter(actual_prices, predicted_prices_scenario2, color='red', label='Scenário 2')

# Scatter plot para o valor médio
mean_prices = np.repeat(np.mean(actual_prices), len(actual_prices))
plt.plot(actual_prices, mean_prices, color='gray', label='Media')

# Plotando a linha da perfeita predição
plt.plot(actual_prices, actual_prices, color='green', linestyle='--', label='Perfect Prediction')

plt.title('Predições vs Atributo Alvo')
plt.xlabel('Preço Real')
plt.ylabel('Preço Predito')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Generate predicted values ranging from 0 to 1
predicted_values = np.linspace(-2, 2, 30)

# Set the expected value to be always 1
expected_value = np.ones_like(predicted_values)

# Compute MAE, MSE, and RMSE for each instance
mae_values = np.abs(expected_value - predicted_values)
mse_values = (expected_value - predicted_values) ** 2
rmse_values = np.sqrt(mse_values)

# Plotting
plt.figure(figsize=(15, 5))

# Plotting MAE
plt.subplot(1, 3, 1)
plt.scatter(predicted_values, mae_values, color='blue')
plt.xlabel('Predicted Values')
plt.ylabel('MAE')
plt.title('MAE vs Predict Values')
plt.grid(True)

# Plotting MSE
plt.subplot(1, 3, 2)
plt.scatter(predicted_values, mse_values, color='green')
plt.xlabel('Predicted Values')
plt.ylabel('MSE')
plt.title('MSE vs Predicted Values')
plt.grid(True)

# Plotting MSE
plt.subplot(1, 3, 3)
plt.scatter(predicted_values, rmse_values, color='red')
plt.xlabel('Predicted Values')
plt.ylabel('RMSE')
plt.title('RMSE vs Predicted Values')
plt.grid(True)
plt.show()

```
