```python
# Célula 1: Imports e Dados (seu Breast Cancer via scikit-learn - shape 569x30)
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
print("Bibliotecas carregadas!")
```

```python
# Célula 2: Carregar e Preparar Dados (X/y definidos!)
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
X = df.drop('target', axis=1)
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"X shape: {X_scaled.shape}, Classes balanceadas: {np.bincount(y)}")
```

```python
# Célula 3: Holdout (seu baseline: ~0.9649)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
model_holdout = Perceptron(random_state=42, max_iter=1000)
model_holdout.fit(X_train, y_train)
acc_holdout = accuracy_score(y_test, model_holdout.predict(X_test))
print(f"**Holdout Accuracy: {acc_holdout:.4f}**")
```

```python
# Célula 4: K-Fold CV (SEM INDENTAÇÃO EXTRA - CORRIGIDO!)
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # <- Linha 1: sem indent!
model_kf = Perceptron(random_state=42, max_iter=1000)
scores_kf = cross_val_score(model_kf, X_scaled, y, cv=kf, scoring='accuracy')
print(f"**5-Fold CV: Média {scores_kf.mean():.4f} (±{scores_kf.std()*2:.4f})**")
print(f"Scores individuais: {scores_kf}")
```

```python
# Célula 5: Stratified K-Fold (melhor pro seu dataset imbalanceado ~37/63%)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores_skf = cross_val_score(model_kf, X_scaled, y, cv=skf, scoring='accuracy')
print(f"**Stratified 5-Fold: Média {scores_skf.mean():.4f} (±{scores_skf.std()*2:.4f})**")
```
